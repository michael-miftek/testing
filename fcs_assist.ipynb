```python
software_names = ['laser1_adc_peak', 'laser1_adc_area', 'laser1_adc_width', 'laser1_photon_total',
                  'laser2_adc_peak', 'laser2_adc_area', 'laser2_adc_width', 'laser2_photon_total',
                  'laser3_adc_peak', 'laser3_adc_area', 'laser3_adc_width', 'laser3_photon_total',
                  'laser4_adc_peak', 'laser4_adc_area', 'laser4_adc_width', 'laser4_photon_total',
                  'laser5_adc_peak', 'laser5_adc_area', 'laser5_adc_width', 'laser5_photon_total',
                  'laser6_adc_peak', 'laser6_adc_area', 'laser6_adc_width', 'laser6_photon_total',
                  'laser7_adc_peak', 'laser7_adc_area', 'laser7_adc_width', 'laser7_photon_total',
                  'time']

hardware_names = ['laser1_abs_area_hw', 'laser1_pos_peak_hw', 'laser1_pos_width_hw', 'laser1_neg_peak_hw', 'laser1_neg_width_hw',
                  'laser2_abs_area_hw', 'laser2_pos_peak_hw', 'laser2_pos_width_hw', 'laser2_neg_peak_hw', 'laser2_neg_width_hw',
                  'laser3_abs_area_hw', 'laser3_pos_peak_hw', 'laser3_pos_width_hw', 'laser3_neg_peak_hw', 'laser3_neg_width_hw',
                  'laser4_abs_area_hw', 'laser4_pos_peak_hw', 'laser4_pos_width_hw', 'laser4_neg_peak_hw', 'laser4_neg_width_hw',
                  'laser5_abs_area_hw', 'laser5_pos_peak_hw', 'laser5_pos_width_hw', 'laser5_neg_peak_hw', 'laser5_neg_width_hw',
                  'laser6_abs_area_hw', 'laser6_pos_peak_hw', 'laser6_pos_width_hw', 'laser6_neg_peak_hw', 'laser6_neg_width_hw',
                  'laser7_abs_area_hw', 'laser7_pos_peak_hw', 'laser7_pos_width_hw', 'laser7_neg_peak_hw', 'laser7_neg_width_hw',
                  'time_hw']


collections_software = {}
collections_hardware = {}
db_client = None
db_software = None
db_hardware = None
```

```python
# def setup_database():
# """
# @brief Function that is used to setup the database

# @details This method will setup the database connection and the collections that are used 
#         in the database. This method will also clear the database of any data that is
#         currently in the database.
        
# @return void
# """
db_client = pymongo.MongoClient("mongodb://localhost:27017/")
db_software = db_client["software"]
db_hardware = db_client["hardware"]

names = db_software.list_collection_names()
collections_software = {}
for name in names:
    db_software.drop_collection(name)
    
try:
    print("Software db: ", db_software)
    for name in software_names:
        collections_software[f'{name}'] = db_software.create_collection(name)
except pymongo.errors.CollectionInvalid:
    print(f"\nException caught in database setup: {e}: {type(e)}. \nSHUTTING DOWN!\n")
    sys.exit()
print("Collections software: ",collections_software)
    
names = db_hardware.list_collection_names()
collections_hardware = {}
for name in names:
    db_hardware.drop_collection(name)
try:
    for name in hardware_names:
        collections_hardware[f'{name}'] = db_hardware.create_collection(name)
except Exception as e:
    print(f"\nException caught in database setup: {e}: {type(e)}. \nSHUTTING DOWN!\n")
    sys.exit()
```

```python
# def load_database(filename):
# """
# @brief Function that is used to load the database from a file

# @details This method will load the database from a file. The file that the database is 
#         loaded from is the filename that is passed to the method. This method will also
#         check to see if the collections that are in the file are in the database. If they 
#         are not in the database, the method will not load the database.
        
# @param filename: The filename that the database is loaded from

# @return void
# """
df = pd.read_csv(f"{csv_file}", low_memory=False)
df = df.drop(df.__len__()-1)

last_line = ''
with open(f"{csv_file}", "r") as file:
    for line in file:
        pass
    last_line = line
    file.close()
    
last_line = last_line.split(" ")
total = int(last_line[0])
load_names = last_line[1:]

load_names2 = []
total2 = None
for i in range(1,len(last_line),1):
    if last_line[i].isdigit():
        load_names2 = last_line[i:]
        break

df2 = None
if load_names2:
    load_names = last_line[1:i]
    total2 = int(load_names2[0])
    load_names2 = load_names2[1:]
    split = total * len(load_names) + len(load_names)
    df2 = df.iloc[split+1:]
    df = df.iloc[:split-1]

# ? Maybe combine the database loads
for i in range(len(load_names)):
    if load_names[i] not in list(collections_software.keys()):
        print("File to load has collections that do not exist in this file, unable to load: ", load_names[i], " not in ", list(collections_software.keys()))
        sys.exit() 
    if (load_names[i] != 'time') and ('laser' not in load_names[i]):
        load_names[i] = 'laser1_' + load_names[i]

    if i == 0:
        continue
    df = df.drop(total*i + (i-1))

for i in range(len(df.columns)):
    if i == 0:
        continue
    elif i == 1:
        df = df.astype({df.columns[i]: int})
    else:
        df = df.astype({df.columns[i]: float})

#NOTE:  This insert into the database walks through the different collections and inserts 
#       the data. The dataframes that correspond to the collections are created by slicing 
#       and saving as tempdf, then all NaNs are dropped from the collection and made into a 
#       dictionary. This dictionary is then inserted into the database.
for i in range(len(load_names)):
    tempdf = df.iloc[(total*i):((total*(i+1))-1)]
    temp = tempdf.dropna(axis=1).to_dict('records')
    collections_software[load_names[i]].insert_many(temp)

if total2:
    names_exist = []
    for i in range(len(load_names2)):
        if load_names2[i] in list(collections_hardware.keys()):
            #NOTE:  Instead of panicking, just skip the collection and move on. If there are 
            #       no collections to load, then return
            names_exist.append(i)
            continue

        if not names_exist:
            print("File to load has collections (hardware) that do not exist, unable to load")
            sys.exit()
        
        if i == 0:
            continue
        df2 = df2.drop(split + total2*i + (i-1))

    for i in range(len(df2.columns)):
        if i == 0:
            continue
        elif i == 1:
            df2 = df2.astype({df.columns[i]: int})
        else:
            df2 = df2.astype({df.columns[i]: float})

    for i in names_exist:
        tempdf = df2.iloc[(total2*i):((total2*(i+1))-1)]
        temp = tempdf.dropna(axis=1).to_dict('records')
        collections_hardware[load_names2[i]].insert_many(temp)

print("Database Loaded")
```

```python
# def close_database():
# """
# @brief Function that is used to close the database connection

# @details This method will close the database connection and the database worker. This
#         method will also drop all of the collections that are in the database. This will
#         clear the database of any data that is currently in the database.
    
# @return void
# """
try:
    names = db_software.list_collection_names()
    for name in names:
        db_software.drop_collection(name)
        
    names = db_hardware.list_collection_names()
    for name in names:
        db_hardware.drop_collection(name)
        
    db_client.close()
    db_client = None
    db_software = None
    db_hardware = None
except Exception as e:
    print(f"Error closing database connection: {e}")
```

# Information Manipulation

These following items will take all the information in the database and align it in the following format

## Laser (#)
### - Channel name (Wavelength nm) - W / Channel name (Wavelength nm) - H / Channel name (Wavelength nm) - A / Channel name (Wavelength nm) - C
#### - Event (Start - Finish)

This will happen for all number of lasers and then can be saved

```python
def sort_laser_params(params):
    def sort_key(item):
        # Split by underscore and extract number and parameter
        laser_num = int(item.split('_')[0].replace('laser', ''))
        param = item.split('_')[1]
        return (laser_num, param)
    
    return sorted(params, key=sort_key)

soft_names = []

names = db_software.list_collection_names()
i = 0
for name in names:
    if db_software.get_collection(name).count_documents({}) == 0:
        continue

    soft_names.append(name)

if 'time' in soft_names:
    soft_names.remove('time')

# Sort and print the result
soft_names = sort_laser_params(soft_names)
print(soft_names)
```

```python
items = {}
for idx in range(soft_names.__len__()):
    cursor = db_software.get_collection(soft_names[idx]).find({}).sort('event', pymongo.ASCENDING)
    items[soft_names[idx]] = pd.DataFrame(list(cursor))
    items[soft_names[idx]].drop(columns=['_id', 'event'], inplace=True)
```

```python
lst = list(items.keys())
print(lst[0])
items[lst[0]].head(5)
```

```python
lst = list(items.keys())
print(len(items[lst[0]].columns))
print(len(channel_names))
print(len([ch for ch in channel_names]))
```

```python
for key in items.keys():
    if 'area' in key:
        sigType = 'A'
    elif 'peak' in key:
        sigType = 'H'
    elif 'width' in key:
        sigType = 'W'
    elif 'photon' in key:
        sigType = 'P'
        items[key][[f'ch {i}' for i in range(48, 60)]] = 0
    items[key].columns = [f'{ch}-{sigType}' for ch in channel_names]
```

```python
numLasers = 0
lasers = []
for key in items.keys():
    if key.split('_')[0] not in lasers:
        numLasers += 1
        lasers.append(key.split('_')[0])

print(numLasers)

lasers = []
for i in range(numLasers):
    keys = [key for key in items.keys() if f'laser{i+1}' in key]
    print(keys)
    lasers.append(
        pd.concat([items[key] for key in keys], axis=1)
    )
```

```python
lasers[0].head(5)
```

```python
organized_cols = []
for ch in channel_names:
    for sign in ['A', 'H', 'W', 'P']:
        organized_cols.append(f'{ch}-{sign}')
print(organized_cols)
```

```python
for i in range(numLasers):
    lasers[i] = lasers[i][organized_cols]
```

```python
for i in range(numLasers):
    lasers[i].to_csv(f'laser{i+1}.csv', index=False)
```
